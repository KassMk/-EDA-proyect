```python
import pandas as pd
url = "https://raw.githubusercontent.com/4GeeksAcademy/data-preprocessing-project-tutorial/main/AB_NYC_2019.csv"
df = pd.read_csv(url)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("df info")
df.info()
print("-"*50)
print("nulos por columna")
print(df.isnull().sum().sort_values(ascending=False) / len(df)*100)
print("-"*50)

df.drop_duplicates(subset=df.columns.difference(["id"]), inplace=True)

columns_drop = ['id', 'host_id', 'host_name', 'last_review', 'name']
df.drop(columns=columns_drop, axis=1, inplace=True)
df.head()
```

```python
import numpy as np

categorical_cols = df.select_dtypes(include='object').columns.tolist()
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print(f"Variables categóricas: {categorical_cols}")
print(f"Variables numéricas:   {numeric_cols}")
print("-"*50)

if 'neighbourhood' in categorical_cols:
    categorical_cols_to_plot = [col for col in categorical_cols if col != 'neighbourhood']

if categorical_cols_to_plot:
    ncols = 2
    nrows = int(np.ceil(len(categorical_cols_to_plot) / ncols))
    fig, axes = plt.subplots(nrows, ncols, figsize=(8 * ncols, 5 * nrows))
    axes = axes.flatten()
    for i, col in enumerate(categorical_cols_to_plot):
        if i < len(axes):
            sns.countplot(ax=axes[i], data=df, x=col, order=df[col].value_counts().index, palette='viridis')
            axes[i].set_title(f'Distribución de {col}', fontsize=14)
            axes[i].tick_params(axis='x', rotation=45)
            axes[i].set_xlabel(col, fontsize=12)
            axes[i].set_ylabel('Conteo', fontsize=12)
        else:
            fig.delaxes(axes[i])

    plt.tight_layout()
    plt.show()

if 'neighbourhood' in df.columns and 'neighbourhood' in categorical_cols:
    print("\n--- Visualización de 'neighbourhood' (Top N Categorías) ---")
    plt.figure(figsize=(15, 7))
    top_n = 20
    top_neighbourhoods = df['neighbourhood'].value_counts().nlargest(top_n).index
    sns.countplot(data=df[df['neighbourhood'].isin(top_neighbourhoods)],
                  x='neighbourhood',
                  order=top_neighbourhoods,
                  palette='viridis')
    plt.title(f'Distribución de neighbourhood (Top {top_n})', fontsize=16)
    plt.xlabel('Vecindario', fontsize=12)
    plt.ylabel('Conteo', fontsize=12)
    plt.xticks(rotation=75, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5)
plt.title("Matriz de Correlación de Variables Numéricas", fontsize=16)
plt.show()
```

```python
price_col_for_viz = 'price'
relevant_categorical_for_price = [col for col in ['neighbourhood_group', 'room_type'] if col in df.columns and col in categorical_cols]

fig, axes = plt.subplots(1, len(relevant_categorical_for_price), figsize=(8 * len(relevant_categorical_for_price), 7))

if len(relevant_categorical_for_price) == 1:
  axes = [axes]

for i, col in enumerate(relevant_categorical_for_price):
  sns.boxplot(ax=axes[i], data=df, x=col, y=price_col_for_viz, palette='Spectral', showfliers=False)
  axes[i].set_title(f'Precio por {col}', fontsize=14)
  axes[i].tick_params(axis='x', rotation=45)
  axes[i].set_xlabel(col, fontsize=12)
  axes[i].set_ylabel(f'Precio', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(data=df, x='longitude', y='latitude', hue=price_col_for_viz, size=price_col_for_viz,
  palette='viridis', sizes=(20, 400), alpha=0.6)
plt.title(f'Mapa de Precios de Airbnb en NYC (Color y Tamaño por Precio)', fontsize=16)
plt.xlabel('Longitud', fontsize=12)
plt.ylabel('Latitud', fontsize=12)
plt.legend(title='Precio', loc='upper right', bbox_to_anchor=(1.2, 1)) # Ajuste de leyenda para que no se superponga
plt.show()
```

```python
other_numeric_for_scatter = [col for col in numeric_cols if col not in ['latitude', 'longitude', 'price']]

ncols_scatter = 3
nrows_scatter = (len(other_numeric_for_scatter) + ncols_scatter - 1) // ncols_scatter
fig, axes = plt.subplots(nrows_scatter, ncols_scatter, figsize=(6 * ncols_scatter, 5 * nrows_scatter))
axes = axes.flatten()

for i, col in enumerate(other_numeric_for_scatter):
  if i < len(axes):
    sns.scatterplot(ax=axes[i], data=df, x=col, y=price_col_for_viz, alpha=0.3, color='blue')
    axes[i].set_title(f'Precio vs. {col}', fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel(f'Precio', fontsize=12)
    axes[i].set_ylim(0, df[price_col_for_viz].quantile(0.99))


plt.tight_layout()
plt.show()
```

```python
df['reviews_per_month'].fillna(0, inplace=True)
df['price_log'] = np.log1p(df['price'])

plt.figure(figsize=(10, 6))
sns.histplot(df['price_log'], kde=True, bins=50, color='purple')
plt.title('Distribución de Log(Precio + 1)', fontsize=16)
plt.xlabel('Log(Precio + 1)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.show()
```

```python
upper_bound_min_nights = df['minimum_nights'].quantile(0.995)
df['minimum_nights_capped'] = df['minimum_nights'].clip(upper=upper_bound_min_nights)

plt.figure(figsize=(10, 6))
sns.histplot(df['minimum_nights_capped'], kde=True, bins=50, color='green')
plt.title('Distribución de Noches Mínimas (Valores Extremos Capados)', fontsize=16)
plt.xlabel('Noches Mínimas (Capado)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.show()
```

```python
df['is_multi_listing_host'] = (df['calculated_host_listings_count'] > 1).astype(int)

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='is_multi_listing_host', y='price_log', palette='crest')
plt.title('Log(Precio) por Anfitrión con Múltiples Listados', fontsize=16)
plt.xlabel('Anfitrión con Múltiples Listados (0=No, 1=Sí)', fontsize=12)
plt.ylabel('Log(Precio + 1)', fontsize=12)
plt.show()

df['has_reviews'] = (df['number_of_reviews'] > 0).astype(int)

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='has_reviews', y='price_log', palette='rocket')
plt.title('Log(Precio) por Si el Listado Tiene Reseñas', fontsize=16)
plt.xlabel('Tiene Reseñas (0=No, 1=Sí)', fontsize=12)
plt.ylabel('Log(Precio + 1)', fontsize=12)
plt.show()
```

```python
columns_to_encode = ['neighbourhood_group', 'room_type']
df_processed = pd.get_dummies(df, columns=columns_to_encode, drop_first=True)
df_processed.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

y = df_processed['price_log']
features_to_exclude = ['price', 'price_log', 'minimum_nights', 'neighbourhood']
X = df_processed.drop(columns=features_to_exclude)

print(f"\nCaracterísticas (X): {X.columns.tolist()}")
print(f"Variable (y): {y.name}")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")
```

```python
import pandas as pd
import os

output_dir = './data/processed'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Directorio '{output_dir}' creado.")
else:
    print(f"Directorio '{output_dir}' ya existe.")

X_train_path = os.path.join(output_dir, 'X_train.csv')
X_train.to_csv(X_train_path, index=False)

X_test_path = os.path.join(output_dir, 'X_test.csv')
X_test.to_csv(X_test_path, index=False)

y_train_path = os.path.join(output_dir, 'y_train.csv')
y_train.to_csv(y_train_path, index=False)

y_test_path = os.path.join(output_dir, 'y_test.csv')
y_test.to_csv(y_test_path, index=False)
```
